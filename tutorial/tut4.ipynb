```python
image = cv2.imread(r"C:\Users\Asus\OPENCV_UDEMY\Computer-Vision-with-Python\DATA\crossword.jpg",0)
```

```python
plt.imshow(image)
```

```python
def show_pic(img):
    fig  = plt.figure(figsize = (15,15))
    ax = plt.subplot(111)
    ax.imshow(img , cmap = 'gray')
    
    
```

```python
show_pic(image)
```

```python
rect,thres = cv2.threshold(image ,180 ,255, type = cv2.THRESH_BINARY )
```

```python
show_pic(thres)
```

```python
thresh2 = cv2.adaptiveThreshold(image , 255 , cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11 , 8)
show_pic(thresh2)
```

```python
blended = cv2.addWeighted(src1 = thres, alpha = 0.6,src2 = thresh2 ,beta = 0.4,gamma =0)
```

```python
import cv2

cv2.imshow('Blended' , blended)
cv2.waitKey() 
    
cv2.destroyAllWindows()
    
```
