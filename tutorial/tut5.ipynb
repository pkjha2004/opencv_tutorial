```python
gray = np.float32(gray_flat_chess)
gray
```

```python
dst = cv2.cornerHarris(src = gray, blockSize = 2, ksize=3, k=0.04)
```

src = source , blocksize = window for calculating , ksize  = change in rate of intesity for any color , k = 0.04 

```python
dst = cv2.dilate(dst , None)
```

dilate which more carefully detect the change in intensity and hence can see the size of the point easily

```python
flat_chess[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(flat_chess)
```
