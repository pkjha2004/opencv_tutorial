```python
import cv2
import time
cap = cv2.VideoCapture(r'C:\Users\Asus\OPENCV_UDEMY\Computer-Vision-with-Python\DATA\My_drawing.mp4')

#making the interacting frame 

#GLobal variables
drawing = False
ix,iy = -1,-1

#Functions
def make_rectangle(event , x,y,flags,param):
    global drawing, ix, iy,frame
    if event == cv2.EVENT_LBUTTONDOWN:
            drawing = True
            img_copy = frame.copy()
            ix,iy = x,y
            
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            img_copy = frame.copy()
            #v2.imshow('Frame', img_copy)
            cv2.rectangle(frame, (ix,iy),(x,y) , (255 , 0 , 0),thickness = 4)
            cv2.imshow('Frame', img_copy)
            
        
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(frame, (ix,iy),(x,y) , (255 , 0 , 0),thickness = 4)
        cv2.imshow('Frame', frame)
        
cv2.namedWindow('Frame')
cv2.setMouseCallback('Frame' ,make_rectangle)


while True:
    ret , frame = cap.read()
    
    if ret:
        cv2.imshow('Frame',frame)
        time.sleep(1/50)
        if cv2.waitKey(1) & 0xFF ==27:
            break
    else:
        break
cap.release()        
cv2.destroyAllWindows()
```
